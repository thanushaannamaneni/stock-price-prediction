```python
!pip install yfinance tensorflow numpy pandas matplotlib scikit-learn -q

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, Conv1D, Flatten
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler

np.random.seed(42)
tf.random.set_seed(42)

ticker = "HPQ"
start_date = "2015-10-27"
end_date = "2025-10-26"
target_date = "2025-10-27"

df = yf.download(ticker, start=start_date, end=end_date)
features = ['Open','High','Low','Close','Volume']
data = df[features].dropna()
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data.values)

time_step = 60
def create_sequences(data_array, time_step):
    X, y = [], []
    for i in range(time_step, len(data_array)):
        X.append(data_array[i-time_step:i])
        y.append(data_array[i, features.index('Close')])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled, time_step)
split = int(len(X)*0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

rnn_model = Sequential([
    SimpleRNN(64, input_shape=(time_step, X.shape[2])),
    Dense(32, activation='relu'),
    Dense(1)
])
rnn_model.compile(optimizer='adam', loss='mse')
es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
rnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, callbacks=[es], verbose=1)

lstm_model = Sequential([
    LSTM(64, input_shape=(time_step, X.shape[2]), return_sequences=True),
    LSTM(32),
    Dense(32, activation='relu'),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
es2 = EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
lstm_model.fit(X_train, y_train, epochs=60, batch_size=32, validation_split=0.1, callbacks=[es2], verbose=1)

cnn_model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(time_step, X.shape[2])),
    Conv1D(32, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(1)
])
cnn_model.compile(optimizer='adam', loss='mse')
es3 = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
cnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, callbacks=[es3], verbose=1)

last_seq = scaled[-time_step:].reshape(1, time_step, X.shape[2])
rnn_pred_scaled = rnn_model.predict(last_seq)[0,0]
lstm_pred_scaled = lstm_model.predict(last_seq)[0,0]
cnn_pred_scaled = cnn_model.predict(last_seq)[0,0]

def inverse_close(val):
    template = scaled[-1].copy().reshape(1,-1)
    template[0, features.index('Close')] = val
    inv = scaler.inverse_transform(template)
    return float(inv[0, features.index('Close')])

rnn_pred = inverse_close(rnn_pred_scaled)
lstm_pred = inverse_close(lstm_pred_scaled)
cnn_pred = inverse_close(cnn_pred_scaled)

print("Predictions for", target_date)
print("RNN:", rnn_pred)
print("LSTM:", lstm_pred)
print("CNN:", cnn_pred)

plt.figure(figsize=(12,6))
plt.plot(pd.to_datetime(df.index[-200:]), data['Close'].values[-200:], label='Actual Close')
next_day = pd.Timestamp(end_date) + pd.Timedelta(days=1)
plt.scatter([next_day], [rnn_pred], label='RNN', marker='x')
plt.scatter([next_day], [lstm_pred], label='LSTM', marker='o')
plt.scatter([next_day], [cnn_pred], label='CNN', marker='s')
plt.legend()
plt.title(f"{ticker} Predicted Close for {target_date}")
plt.show()

rnn_model.save("rnn_model.h5")
lstm_model.save("lstm_model.h5")
cnn_model.save("cnn_model.h5")
print("Models saved.")
```
